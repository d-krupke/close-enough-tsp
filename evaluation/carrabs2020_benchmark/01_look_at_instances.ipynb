```python
from cetsp_bnb2.core import (Instance, Circle, Point, plot_instance)
```

```python
import gzip
import json

with open("00_instances/benchmark.json.gz", "rb") as f:
    instances = json.loads(gzip.decompress(f.read()).decode("utf-8"))

instance_names = list(instances.keys())
```

```python
import matplotlib.pyplot as plt
import random
for i in range(5):
    instance_name = random.choice(instance_names)
    instance = Instance([Circle(Point(d["x"], d['y']), d["r"]) for d in instances[instance_name]["disks"]])
    fig, ax = plt.subplots(figsize=(10, 10))
    plt.title(instance_name)
    plot_instance(ax, instance)
    plt.show()
```

```python
from cetsp_bnb2.core import optimize, plot_solution

instance_name = "concentricCircles2"
instance = Instance([Circle(Point(d["x"], d['y']), d["r"]) for d in instances[instance_name]["disks"]])

fig, ax = plt.subplots(figsize=(10, 10))
plt.title(instance_name)
plot_instance(ax, instance)
plt.show()

sol = optimize(instance, optimality_gap=0.001, feasibility_tol=0.0001, use_stronger_lb=False)
fig, ax = plt.subplots(figsize=(10, 10))
plt.title(instance_name)
plot_solution(ax, instance, sol[0])
plt.show()
```
